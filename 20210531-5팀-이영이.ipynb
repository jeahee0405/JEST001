```python
# 예제1. 편의점 재고 관리
# 편의점 재고 관리 프로그램을 작성해보자
# 엔터를 입력할 때까지 반복하여 사용자로부터 물건이름과 개수를 입력받는다. *while문을 사용하여 반복하고, 딕셔너리로 key를 물건이름
# 만약 물건 이름이 존재하지 않으면 추가하고, 존재하면 물건 개수를 더한다. *if 문
# 입력이 끝나면 재고 현황을 물건이름의 오름차순으로 정렬하여 출력한다. *딕셔너리 전체를 물건이름으로 정렬하여 for문
```

```python
product = {}
# 물건이름 = key(name), 개수 = value(num)
while True : #무한반복
    name = input('물건이름:') #신규이름을 받고
    if name == '' : #반복문중단 조건
        break
    num = int(input('개수:'))
    if name in product.keys() : #있던거면
        product[name] += num #개수를 더하고
    else :
        product[name] = num #없으면 value 값, 내가 생각 못한 부분
print(product)

#오름차순 정렬
for key,value in sorted(product.items()) : #딕셔너리 전체를 물건 이름으로 정렬하는 거
    print(key, value, end= ' ')
```

```python
# 집합 연산
# 합집합 a.union(b) a|b
# 교집합 a.intersction(b) a&b
# 차집합 a.difference(b) a-b
```

```python
sweety = {'조청유과','칙촉','허니버터칩','새우깡'}
salty = {'새우깡','무뚝뚝','자가비'}

# 합집합
print('합집합')
print(sweety.union(salty))
print(sweety | salty)
print('-' * 40)
print('교집합')
print(sweety.intersection(salty))
print(sweety&salty)
print('-' * 40)
print('차집합')
print(sweety.symmetric_difference(salty)) #이건뭐지?
print(sweety.difference(salty))
print(sweety-salty)
```

```python
# 예제. 금액에 따른 할인율 계산하기
# 1만원 이상 ~ 5만원 미만 = 5% 할인
# 5만원 이상 ~ 10만원 미만 = 7.5% 
# 10만원 이상 = 10%
```

```python
#코드짜는 것보다 할인율 사칙연산이 더어렵다 ㅋㅋㅋㅋ
price = int(input("구매한 물건 가격은 얼마인가요?: "))
sale = ()

if price >= 100000 :
    sale = (10/100)
elif 50000 <= price < 100000 :
    sale = (7.5/100)
elif 10000 <= price < 50000 :
    sale = (5/100) 
else :
    print("할인혜택이 없습니다.")

pay = int(price*sale)

print(f"구매가 {price}원 이고, 할인가{pay}원으로 할인율은 {sale}%, 최종 결제금액은 {(price-pay)}원")
```

```python
# 사칙연산 프로그램 문제분석
# 두 수와 사칙연산기호(+,-,*,/)를 입력 받아 연산 기호에 따라 연산 결과를 출력하는 프로그램을 작성하세요

# 결과화면
# 숫자1 : 1
# 숫자2 : 3
# 연산기호 : +
# 1 + 3 = 4

num1 = int(input('첫번째 수를 입력하세요:'))
num2 = int(input('두번쨰 수를 입력하세요:'))
sign = input('연산기호를 선택하세요(+,-,*,/): ')

print(f'첫번째 수 = {num1}')
print(f'두번째 수 = {num2}')
print(f'연산기호 = {sign}')
print('-' * 40)
print(f'{num1} {sign} {num2} = {}') # 으아 오후5시 되니 머리가 뻗는다
```

```python
# 사칙연산 선생님코드

num1 = int(input('숫자 1 : '))
num2 = int(input('숫자 2 : '))
operator = input('연산기호(+,-,*,/) : ')
result = 0

if operator == '+':
    result = num1 + num2
elif operator == '-' :
    result = num1 - num2
elif operator == '*' :
    result = num1 * num2
elif operator == '/' :
    result = num1 / num2 #0으로 나누명 오류남

print(f'{num1} {operator} {num2} = {result}')
```

```python
# 5팀 신동열님 : 10번 주사위를 던졌을 때 승패 비율까지 구하기

win_1=[0]
win_2=[0]
lose_1=[0]
lose_2=[0]
import random
for i in range(10):
    no1=random.randint(1,6)
    no2=random.randint(1,6)
    if no1>no2:
        print(f'주사위1:{no1},주사위2:{no2} 주사위 1이 이겼습니다')
        win_1[0]+=1
        lose_2[0]+=1
    elif no1<no2:
        print(f'주사위1:{no1},주사위2:{no2} 주사위 2가 이겼습니다')
        win_2[0]+=1
        lose_1[0]+=1
    else :
        print(f'주사위1:{no1},주사위2:{no2} 주사위 수가 같습니다.')
print(f'''주사위1 {win_1}승{lose_1}패
주사위2 {win_2}승 {lose_2}패''')
```
