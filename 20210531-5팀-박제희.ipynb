```python
#집한(set)>연산
sweety={'빼빼로','칸초','허니버터'}
salty={'허니버터','포카칩','새우깡'}

#합집합
print(sweety.union(salty))
print(sweety|salty)
#교집합
print(sweety.intersection(salty))
print(sweety&salty)
#차집합
print(sweety.difference(salty))
print(sweety-salty)
print(salty.difference(sweety))
print(salty-sweety)
```

```python
#예제>점수에 따른 등급 출력하기(점수처리)

grade=' '
score=int(input('점수를 입력하세요'))
if score>=90:
    grade='A'
elif score>=80:
    grade='B'
elif score>=70:
    grade='C'
elif score>=60:
    grade='D'
else:
    grade='F'
print(f'등급:{grade}')
```

```python
#예제>금액에 따른 할인율 계산하기

buy=int(input('물건 구매가:'))
if buy>=100000:
    percentage='10'
elif buy>=50000:
    percentage='7.5'
elif buy>=10000:
    percentage='5'
dc_price = buy*(percentage/100)

print(f'구매가:{buy}원')
print(f'할인율:{percentage}%')
print(f'할인금액:{int(dc_price)}원')
print(f'지불금액:{int(buy-dc_price)}')
```

```python
#주사위게임

import random
no1=random.randint(1,6)
print('주사위1:',no1)
no2=random.randint(1,6)
print('주사위2:', no2)

if no1>no2:
    print(f'주사위 1이 이겼습니다')
elif no1<no2:
    print(f'주사위 2가 이겼습니다')
else :
    print(f'주사위 수가 같습니다.')
```

```python
win_1=[0]
win_2=[0]
lose_1=[0]
lose_2=[0]
import random
for i in range(10):
    no1=random.randint(1,6)
    no2=random.randint(1,6)
    if no1>no2:
        print(f'주사위1:{no1},주사위2:{no2} 주사위 1이 이겼습니다')
        win_1[0]+=1
        lose_2[0]+=1
    elif no1<no2:
        print(f'주사위1:{no1},주사위2:{no2} 주사위 2가 이겼습니다')
        win_2[0]+=1
        lose_1[0]+=1
    else :
        print(f'주사위1:{no1},주사위2:{no2} 주사위 수가 같습니다.')
print(f'''주사위1 {win_1}승{lose_1}패
주사위2 {win_2}승 {lose_2}패''')
```
