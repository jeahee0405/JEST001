딕셔너리 예제

```python
#편의점 재고관리 
#엔터를 입력할때까지 반복하여 사용자로부터 물건이름과 개수를 입력 ( while문 사용하여 반복, 물건이름/개수 ->쌍, key는 물건이름)
#물건이름이 존재하지않으면 추가하고, 존재하면 물건개수를 더한다 (if ~ else 구조)
#입력이 끝나면 재고 현황을 물건이름의 오름차순으로 정렬하여 출력한다 ( for문을 돌려서 출력 )
d_li = {}
#물건이름 : stuff, 개수 num
while True:
  stuff = input('물건이름:')
  if stuff=='':
    break 
  num = int(input('개수:'))
  if stuff in d_li.keys():
     d_li[stuff] += num
  else:
    d_li[stuff] = num
print(d_li)
#오름차순정렬
for key,value in sorted(d_li.items()):
    print(key,value,end=' ')
```

# 집합(set)연산


*   합집합 union a|b
*   교집합 intersction a&b
*   차집합 difference a-b



```python
sweet = {'빼빼로','칸초','허니버터'}
salt = {'허니버터','포카칩','새우깡'}
#합집합
print(sweet.union(salt))
print(sweet|salt)
#교집합 
print(sweet.intersection(salt))
print(sweet&salt)
#차집합
print(sweet.difference(salt))
print(sweet-salt)
```

```python
#예제2 로그인 프로그램 
# 사용자로부터 아이디/패스워드를 받아서 저장된 아이디/패스워드와 일치여부 검사 후 로그인/로그인 실패 출력하는 프로그램작성 
# 아이디:ID1234, 패스워드:PWD1234 
user_id = 'id1234'
user_pw = 'pwd1234'
id = input('아이디 : ')
pw = input('PW :')

if user_id == id and user_pw == pw:
    print('로그인 성공')
else:
    print('로그인 실패')
```

```python
#예제3 사칙연산 프로그램 - 문제분석 
# > 두 수와 사칙연산기호(+,-,*,/)를 입력받아 연산기호에 따라(조건문 if) 연산결과를 출력하시오 
num1 = int(input('숫자1 : '))
num2 = int(input('숫자2 : '))
num3 = input('연산기호(+,-,*,/) : ')
if num3 == '+':
    print(num1+num2)
elif num3=='-':
    print(num1-num2)
elif num3=='*':
    print(num1*num2)
else:
    print(float(num1/num2))
```
