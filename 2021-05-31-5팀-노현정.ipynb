#집합 연산
합집합 union :shift+원화기호 ,
교집합 intersction :&,
차집합 difference: -



```python
sweety = {'chocolate', 'jelly', 'caramel'}
salty = {'potatochip', 'honeybutterchip', 'chichenpop', 'caramel'}

# 합집합
print(sweety.union(salty))
print(sweety|salty)

# 교집합
print(sweety.intersection(salty))
print(sweety&salty)

# 차집합
print(sweety.difference(salty))
print(sweety-salty)
print(salty.difference(sweety))
print(salty-sweety)
```
