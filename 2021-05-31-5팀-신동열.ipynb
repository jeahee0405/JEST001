```python
점수=int(input('점수를 입력하시오:'))

if 점수>=90:
    print('등급:A')
elif 80<=점수<=89:
    print('등급:B')
elif 70<=점수<=79:
    print('등급:C')
elif 60<=점수<=69:
    print('등급:D')
else:
    print('등급:F')
```

```python
물건구매가=int(input('물건 구매가를 입력하세요.:'))
if 10000<=물건구매가<50000:
    할인율=5
elif 50000<=물건구매가<100000:
    할인율= 7.5
else:
    할인율= 10

할인금액=int((물건구매가)*(할인율/100))
지불금액=물건구매가-할인금액
print(f'''물건구매가:{물건구매가}원     
할인율:{할인율}%
할인금액:{할인금액}원
지불금액:{지불금액}원''')
```

```python
no1=int(input('숫자1:'))
no2=int(input('숫자2:'))
nom=input('연산기호(+,-,*,/):')

if nom== '+':
   col=no1+no2
elif nom== '-':
    col=no1-no2
elif nom== '*':
    col=no1*no2
elif nom== '/':
    col=no1/no2

print(f'{no1} {nom} {no2} = {col}')
```

```python
import random

no1=random.randint(1,6)
no2=random.randint(1,6)


print('주사위1:',no1)
print('주사위2:', no2)

if no1>no2:
    print(f'주사위 1이 이겼습니다')
elif no1<no2:
    print(f'주사위 2가 이겼습니다')
else :
    print(f'주사위 수가 같습니다.')
```

```python
win_1=[0]
win_2=[0]
lose_1=[0]
lose_2=[0]

import random

for i in range(10):
    no1=random.randint(1,6)
    no2=random.randint(1,6)
    if no1>no2:
        print(f'주사위1:{no1},주사위2:{no2} 주사위 1이 이겼습니다')
        win_1[0]+=1
        lose_2[0]+=1
    elif no1<no2:
        print(f'주사위1:{no1},주사위2:{no2} 주사위 2가 이겼습니다')
        win_2[0]+=1
        lose_1[0]+=1
    else :
        print(f'주사위1:{no1},주사위2:{no2} 주사위 수가 같습니다.')

print(f'''주사위1: {win_1}승{lose_1}패
주사위2: {win_2}승 {lose_2}패''')
```
